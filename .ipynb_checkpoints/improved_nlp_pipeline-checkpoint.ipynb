# 📓 NLP Pipeline & Preprocessing Notebook

---

## 📌 Introduction to NLP

**Natural Language Processing (NLP)** is a subfield of Artificial Intelligence that enables machines to understand, interpret, and generate human language in a meaningful and useful way.

### 🧠 What is an NLP Pipeline?

An **NLP Pipeline** is a structured sequence of steps that transforms raw, unstructured text into a clean, numerical format suitable for machine learning models and deep learning algorithms.

### 📊 Common NLP Use Cases

- 💭 **Sentiment Analysis**: Determining emotional tone (positive/negative/neutral)
- 🚫 **Spam Detection**: Filtering unwanted emails and messages
- 🏷️ **Text Classification**: Categorizing documents into predefined categories
- 🤖 **Chatbots**: Building conversational AI systems
- 🔍 **Information Retrieval**: Searching and ranking relevant documents
- 📄 **Machine Translation**: Converting text between languages
- 🎯 **Named Entity Recognition**: Identifying proper nouns and entities

---

## 🔄 The NLP Pipeline Overview

```
Raw Text → Tokenization → Normalization → Vectorization → ML Model → Output
```

Each stage is crucial for preparing text data for machine learning.

# 🔀 Stage 1: Tokenization

## Definition

**Tokenization** is the process of breaking down text into smaller meaningful units called **tokens**. Tokens can be words, sentences, characters, or subwords depending on the tokenization method used.

### Why Tokenization Matters
- Converts text into discrete units that models can process
- Reduces data complexity
- Serves as the foundation for all downstream NLP tasks

### Types of Tokenization
1. **Word Tokenization**: Splits text into individual words
2. **Sentence Tokenization**: Splits text into sentences
3. **Character Tokenization**: Splits text into individual characters
4. **Regex Tokenization**: Uses regular expressions for custom splits

---

## 1️⃣ Word Tokenization

**Purpose**: Splits text into individual words.

**Use Case**: Basic text processing for bag-of-words models, keyword extraction.

**Example**:
```
Input: "He has 35$"
Output: ['He', 'has', '35', '$']
```

````python
import nltk
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize

text = "He has 35$"
tokens = word_tokenize(text)

print(f"Original text: {text}")
print(f"Tokens: {tokens}")
print(f"Number of tokens: {len(tokens)}")
````

## 2️⃣ Sentence Tokenization

**Purpose**: Splits text into individual sentences.

**Use Case**: Document analysis, paragraph-level sentiment analysis, dialogue systems.

**Example**:
```
Input: "Adam was here yesterday? Why didn't you tell me?"
Output: ['Adam was here yesterday?', 'Why didn't you tell me?']
```

````python
from nltk.tokenize import sent_tokenize

text = "Adam was here yesterday? Why didn't you tell me? This is important!"
sentences = sent_tokenize(text)

print(f"Original text: {text}\n")
print("Sentences:")
for i, sent in enumerate(sentences, 1):
    print(f"  {i}. {sent}")
````

## 3️⃣ Character Tokenization

**Purpose**: Splits text into individual characters.

**Use Case**: Language modeling at character level, handling misspellings, non-English text.

**Example**:
```
Input: "NLP"
Output: ['N', 'L', 'P']
```

````python
text = "NLP"
char_tokens = list(text)

print(f"Original text: {text}")
print(f"Character tokens: {char_tokens}")
````

## 4️⃣ Regex Tokenization

**Purpose**: Uses regular expressions to extract tokens based on custom patterns.

**Use Case**: Domain-specific tokenization, extracting alphanumeric tokens, handling special formats.

**Example**:
```
Input: "NLP is great, AI is the future!"
Pattern: Extract only words (\\b\\w+\\b)
Output: ['NLP', 'is', 'great', 'AI', 'is', 'the', 'future']
```

````python
import re

text = "NLP is great, AI is the future!"
tokens = re.findall(r"\b\w+\b", text)

print(f"Original text: {text}")
print(f"Regex tokens: {tokens}")
print(f"\nExplanation: Pattern \\b\\w+\\b extracts word boundaries and alphanumeric characters")
````

# 📝 Stage 2: Normalization

## Definition

**Normalization** is the process of standardizing text to reduce its dimensionality and improve model performance. It involves cleaning and transforming text into a consistent format.

### Why Normalization Matters
- Reduces redundancy (e.g., "running", "runs", "run" are treated as the same concept)
- Improves model accuracy by reducing feature space
- Handles inconsistencies in text data
- Reduces noise from punctuation, numbers, and special characters

### Common Normalization Techniques
1. **Lowercasing**: Convert all text to lowercase
2. **Stemming**: Reduce words to their root form (rule-based)
3. **Lemmatization**: Reduce words to their base form (dictionary-based)
4. **Stopword Removal**: Remove common, non-informative words
5. **Number Handling**: Remove or replace numbers
6. **Repeated Character Handling**: Fix repeated characters
7. **Contraction Expansion**: Expand contractions (e.g., "don't" → "do not")

---

## 1️⃣ Lowercasing

**Purpose**: Converts all uppercase letters to lowercase.

**Benefit**: Treats "Hello", "HELLO", and "hello" as the same word.

**Example**:
```
Input: "Hello NLP"
Output: "hello nlp"
```

````python
text = "Hello NLP World"
lowercased = text.lower()

print(f"Original: {text}")
print(f"Lowercased: {lowercased}")
````

## 2️⃣ Stemming

**Purpose**: Reduces words to their root form using rule-based algorithms.

**Algorithm**: Porter Stemmer is the most common algorithm that removes common suffixes.

**Pros**: Fast, works well for most languages

**Cons**: May produce non-dictionary words (e.g., "studi" from "studies")

**Example**:
```
running → run
played → play
studies → studi
```

````python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
words = ["running", "played", "studies", "flies", "agreed"]

print("Stemming Examples:")
print("-" * 40)
for word in words:
    stem = stemmer.stem(word)
    print(f"{word:15} → {stem}")
````

## 3️⃣ Lemmatization

**Purpose**: Reduces words to their base/dictionary form (lemma) using linguistic knowledge.

**Algorithm**: Uses WordNet lexical database and part-of-speech (POS) tags.

**Pros**: Produces valid dictionary words, more linguistically accurate

**Cons**: Slower than stemming, requires POS tagging

**Example**:
```
running → run
better → good
best → best (no change needed)
```

### Stemming vs Lemmatization
| Aspect | Stemming | Lemmatization |
|--------|----------|---------------|
| Approach | Rule-based | Dictionary-based |
| Output | May not be valid words | Always valid words |
| Speed | Fast | Slower |
| Accuracy | Lower | Higher |
| Use Case | Information retrieval | Sentiment analysis |

````python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print("Lemmatization Examples:")
print("-" * 50)

# Verb lemmatization
print("Verbs:")
print(f"  running → {lemmatizer.lemmatize('running', pos='v')}")
print(f"  played → {lemmatizer.lemmatize('played', pos='v')}")

# Adjective lemmatization
print("\nAdjectives:")
print(f"  better → {lemmatizer.lemmatize('better', pos='a')}")
print(f"  greatest → {lemmatizer.lemmatize('greatest', pos='a')}")

# Noun lemmatization
print("\nNouns:")
print(f"  studies → {lemmatizer.lemmatize('studies', pos='n')}")
````

## 4️⃣ Stopword Removal

**Purpose**: Removes common, non-informative words that don't add much value to analysis.

**Common Stopwords**: "the", "is", "at", "which", "on", "a", "an", "am", "to", etc.

**Benefit**: Reduces dimensionality and noise in the dataset.

**Caveat**: May remove important words in some contexts (e.g., "not" is crucial for sentiment analysis).

**Example**:
```
Input: ["I", "am", "going", "to", "school"]
Output: ["I", "going", "school"]
```

````python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

text = ["I", "am", "going", "to", "school"]
filtered = [w for w in text if w.lower() not in stop_words]

print(f"Original tokens: {text}")
print(f"Filtered tokens: {filtered}")
print(f"\nRemoved stopwords: {[w for w in text if w.lower() in stop_words]}")
print(f"\nTotal English stopwords: {len(stop_words)}")
````

## 5️⃣ Removing/Normalizing Numbers

**Purpose**: Removes or replaces numerical values to reduce feature space.

**Options**:
- Remove all numbers completely
- Replace numbers with a placeholder like `<NUM>`
- Normalize numbers (e.g., "1000000" → "1M")

**Use Case**: When numbers are not informative for the task.

**Example**:
```
Input: "I have 3 cats and 10 dogs"
Output: "I have <NUM> cats and <NUM> dogs"
```

````python
import re

text = "I have 3 cats and 10 dogs. My phone number is 555-1234."
normalized = re.sub(r"\d+", "<NUM>", text)

print(f"Original: {text}")
print(f"Normalized: {normalized}")
````

## 6️⃣ Handling Repeated Characters

**Purpose**: Fixes repeated characters common in social media text and informal writing.

**Benefit**: Normalizes expressions like "soooooo" to "so".

**Example**:
```
Input: "soo good heyyyy"
Output: "so good hey"
```

````python
import re

text = "soo good heyyyy!!!! I loooove this!!!"
# Replace 3 or more repeated characters with just 1
normalized = re.sub(r"(.)\1{2,}", r"\1", text)

print(f"Original: {text}")
print(f"Normalized: {normalized}")
````

## 7️⃣ Contraction Expansion

**Purpose**: Expands contractions to their full form.

**Benefit**: Ensures consistency and improves tokenization.

**Common Contractions**:
- "don't" → "do not"
- "can't" → "cannot"
- "I'm" → "I am"
- "it's" → "it is"

**Example**:
```
Input: "I'm happy because I don't care"
Output: "I am happy because I do not care"
```

````python
contractions_dict = {
    "don't": "do not",
    "i'm": "i am",
    "can't": "cannot",
    "won't": "will not",
    "it's": "it is",
    "they're": "they are"
}

text = "I'm happy because I don't care. They're coming but it's too late."

# Case-insensitive contraction expansion
for contraction, expansion in contractions_dict.items():
    text = text.replace(contraction, expansion)

print(f"Expanded text: {text}")
````

# 🔢 Stage 3: Vectorization

## Definition

**Vectorization** (also called **Feature Extraction**) is the process of converting text into numerical vectors that machine learning models can understand and process.

### Why Vectorization Matters
- ML models work with numbers, not text
- Captures semantic meaning and word relationships
- Enables mathematical operations on text

### Common Vectorization Techniques
1. **Bag of Words (BoW)**: Counts word occurrences
2. **TF-IDF**: Weights words by importance
3. **Word Embeddings**: Dense vector representations

---

## 1️⃣ Bag of Words (BoW)

### Definition

**Bag of Words** represents text as an unordered collection of words and their frequencies. Word order is not preserved.

### Two Variants

#### A) Binary Bag of Words
- Each word has a value of 1 (present) or 0 (absent)
- Ignores word frequency

#### B) Frequency Bag of Words (Count Vectorization)
- Each word has a value equal to its frequency in the document
- More informative than binary BoW

### Pros and Cons
| Aspect | Pros | Cons |
|--------|------|------|
| Simplicity | Easy to understand and implement | Loses word order information |
| Speed | Fast computation | Cannot capture context |
| Sparsity | Works well for small datasets | High dimensionality for large vocabularies |

````python
from sklearn.feature_extraction.text import CountVectorizer

corpus = ["I love NLP", "NLP is fun"]

# Binary BoW
print("=" * 50)
print("BINARY BAG OF WORDS")
print("=" * 50)
vectorizer_binary = CountVectorizer(binary=True)
X_binary = vectorizer_binary.fit_transform(corpus)

print(f"\nVocabulary: {vectorizer_binary.get_feature_names_out()}")
print(f"\nBinary BoW Matrix:")
print(X_binary.toarray())
print(f"\nInterpretation:")
for i, doc in enumerate(corpus):
    print(f"  Doc {i+1} ('{doc}'): {X_binary.toarray()[i]}")
````

````python
from sklearn.feature_extraction.text import CountVectorizer

corpus = ["I love NLP and also love deep learning", "NLP is fun"]

# Frequency BoW
print("=" * 60)
print("FREQUENCY BAG OF WORDS (COUNT VECTORIZATION)")
print("=" * 60)
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)

print(f"\nVocabulary: {vectorizer.get_feature_names_out()}")
print(f"\nFrequency BoW Matrix:")
print(X.toarray())
print(f"\nInterpretation:")
for i, doc in enumerate(corpus):
    print(f"  Doc {i+1}: {X.toarray()[i]}")
    
print(f"\nNote: The word 'love' appears 2 times in Doc1, so its value is 2")
````

## 2️⃣ TF-IDF (Term Frequency – Inverse Document Frequency)

### Definition

**TF-IDF** is a numerical statistic that reflects the importance of a word in a document within a corpus.

### Formula

```
TF-IDF(w, d) = TF(w, d) × IDF(w)

Where:
  TF(w, d) = (Frequency of word w in document d) / (Total words in d)
  IDF(w) = log(Total documents / Documents containing word w)
```

### How It Works

- **TF (Term Frequency)**: How often a word appears in a specific document
- **IDF (Inverse Document Frequency)**: How rare/unique a word is across all documents
- **High TF-IDF**: Word is frequent in a document but rare across documents (important!)
- **Low TF-IDF**: Word is common across all documents (not important)

### Example:
```
Corpus:
  Doc1: "I love NLP"
  Doc2: "NLP is fun"

TF-IDF values:
  "love" has high TF-IDF (appears in only 1 doc)
  "NLP" has lower TF-IDF (appears in both docs)
  "is" has low TF-IDF (common word)
```

````python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = ["I love NLP", "NLP is fun"]

print("=" * 60)
print("TF-IDF VECTORIZATION")
print("=" * 60)

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)

print(f"\nVocabulary: {vectorizer.get_feature_names_out()}")
print(f"\nTF-IDF Matrix:")
print(X.toarray())
print(f"\nInterpretation:")
for i, doc in enumerate(corpus):
    print(f"\n  Doc {i+1} ('{doc}'):")
    for word, score in zip(vectorizer.get_feature_names_out(), X.toarray()[i]):
        if score > 0:
            print(f"    {word}: {score:.4f}")
````

## 3️⃣ Word Embeddings

### Definition

**Word Embeddings** are dense vector representations of words in a continuous vector space where semantically similar words are close to each other.

### Key Concepts

- **Dense Vectors**: Unlike BoW, embeddings use compact fixed-size vectors
- **Semantic Similarity**: Words with similar meanings have similar vectors
- **Dimensionality**: Typically 50-300 dimensions (vs thousands for BoW)

### Popular Methods

1. **Word2Vec**: Google's algorithm (Skip-gram, CBOW)
2. **GloVe**: Global Vectors for Word Representation
3. **FastText**: Facebook's method handling subword information
4. **ELMo**: Contextual embeddings
5. **BERT**: Bidirectional Encoder Representations from Transformers

### Average Word Embedding

A simple approach to get sentence/document embeddings:
1. Get embeddings for each word
2. Average all word vectors

````python
import numpy as np

# Simulated word embeddings (in practice, use pre-trained models like Word2Vec)
embeddings = {
    "i": [1, 0, 0],
    "love": [0, 1, 0],
    "nlp": [0, 0, 1],
    "is": [0.5, 0.5, 0],
    "fun": [0, 0.8, 0.2]
}

def get_average_embedding(sentence):
    """Get average embedding for a sentence."""
    words = sentence.lower().split()
    vectors = [embeddings[w] for w in words if w in embeddings]
    if not vectors:
        return None
    return np.mean(vectors, axis=0)

print("=" * 60)
print("AVERAGE WORD EMBEDDING")
print("=" * 60)

sentences = ["I love NLP", "NLP is fun"]

for sent in sentences:
    emb = get_average_embedding(sent)
    print(f"\nSentence: '{sent}'")
    print(f"Average Embedding: {emb}")
    print(f"Embedding dimension: {len(emb)}")
````
