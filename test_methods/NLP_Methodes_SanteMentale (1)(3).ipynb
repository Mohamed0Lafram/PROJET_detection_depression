## 1. NLP Pipeline

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# télécharger toutes les ressources nécessaires
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = str(text).lower()
    
    # supprimer urls
    text = re.sub(r"http\S+", "", text)
    
    # supprimer caractères spéciaux
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    
    # tokenization
    tokens = word_tokenize(text)
    
    # supprimer stopwords
    tokens = [word for word in tokens if word not in stop_words]
    
    return tokens

print("Texte original :")
print(df['statement'][0])

print("\nTexte après nettoyage :")
print(clean_text(df['statement'][0]))
```


Ici on prépare les données :

- suppression ponctuation
- suppression stopwords
- tokenisation

Le texte devient plus propre avant l'entraînement.


## 2. Bag of Words

```python

from sklearn.feature_extraction.text import CountVectorizer

sample=df['statement'].head(5)

vectorizer=CountVectorizer()
X=vectorizer.fit_transform(sample)

import pandas as pd
bow_df=pd.DataFrame(
    X.toarray(),
    columns=vectorizer.get_feature_names_out()
)

bow_df
```


Bag of Words compte simplement les mots.

Si "anxiety" apparaît souvent dans les textes Anxiety → le modèle apprend ce pattern.


## 3. TF-IDF

```python

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf=TfidfVectorizer()

X_tfidf=tfidf.fit_transform(sample)

tfidf_df=pd.DataFrame(
    X_tfidf.toarray(),
    columns=tfidf.get_feature_names_out()
)

tfidf_df.head()
```


TF-IDF donne plus d’importance aux mots rares mais importants.

Exemple :

- "the" → faible poids
- "suicide" → poids élevé

Très utile pour ton dataset.


## 4. Average Word Embedding

```python

import gensim.downloader as api
import numpy as np

model=api.load("glove-wiki-gigaword-50")

def avg_embedding(text):
    words=text.lower().split()
    vectors=[model[w] for w in words if w in model]

    if len(vectors)==0:
        return np.zeros(50)

    return np.mean(vectors,axis=0)

vec=avg_embedding(df['statement'][0])

print("Taille du vecteur:", len(vec))
print(vec[:10])
```


Chaque mot devient un vecteur mathématique.

Cette méthode comprend mieux les relations entre :

- sad
- depressed
- unhappy
